```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"] # 设置默认字体为 SimHei，以便支持中文
mpl.rcParams["axes.unicode_minus"] = False # 设置为False，以便图形中负号正常显示。
 
plt.xlabel("箱子编号")
plt.ylabel("箱子重量(kg)")

# 构建柱状图
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 1, 4, 5, 8, 9, 7, 2]
plt.bar(x, y, tick_label=["a", "b", "c", "d", "e", "f", "g", "h"], hatch="/", color="c")

plt.show()
```
