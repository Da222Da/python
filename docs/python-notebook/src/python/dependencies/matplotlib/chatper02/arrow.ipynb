```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"] # 设置默认字体为 SimHei，以便支持中文
mpl.rcParams["axes.unicode_minus"] = False # 设置为False，以便图形中负号正常显示。
fig, ax = plt.subplots() 


# 标题
plt.title("用箭头来表示二维向量")

# 设置x轴和y轴的范围
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# 绘制点
x = 0
y = 0
n = 3
m = 2
ax.scatter(x, y, marker='x', color='red')
ax.scatter(n, m, marker='x', color='red')
ax.text(x + 0.2, y - 0.2, f"({x}, {y})")
ax.text(n + 0.2, m - 0.2, f"({n}, {m})")
ax.arrow(x, y, n - 0.2, m - 0.15, head_width=0.2, head_length=0.2, fc="black", ec="black")

# 辅助虚线
ax.axvline(x=x, linestyle='--')
ax.axvline(x=n, linestyle='--')

# 显示图形
plt.show()
```
