```python
# 如何在图形中支持中文字符？

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"] # 设置默认字体为 SimHei，以便支持中文
mpl.rcParams["axes.unicode_minus"] = False # 设置为False，以便图形中负号正常显示。

fig, ax = plt.subplots() 

# 标题
plt.title("二维空间结构示意图")

plt.xlim(-5, 5)
plt.ylim(-5, 5)

# 绘制点
x = 0
y = 0

ax.scatter(x, y)

# 带箭头的文字说明
plt.annotate("参考点：原点", 
        c="r", 
        weight="bold", 
        xy=(x, y),  # 被注释图形内容的位置坐标
        xytext=(x + 0.8, y + 1.2),  # 注释文本的位置坐标
        arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="r") # 被注释内容的箭头属性字典
)

# 显示图形
plt.show()
```
