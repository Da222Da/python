# 二维向量

二维向量就是二维空间中相对于原点的某个点`(x, y)`。主要内容如下：
- 什么是二维空间？
- 如何表示二维向量？   
- 示例：绘制恐龙

## 什么是二维空间？

二维世界如同一页纸或者计算机屏幕一样扁平。在数学语言中，这种扁平的二维空间称为**平面**。

你可以把这个平面想象是由无数个点组合而成的，为了描述这些点在平面中的位置，我在平面中定义了一个参考点`原点`，然后，通过垂直位置`y`和水平位置`x`这两个信息来描述某个点在二维世界中的具体位置。
 
总的来说，**二维空间 = 原点(0, 0) + n 个坐标点(n, m) + 辅助测量的坐标轴x, y**


```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"] # 设置默认字体为 SimHei，以便支持中文
mpl.rcParams["axes.unicode_minus"] = False # 设置为False，以便图形中负号正常显示。

fig, ax = plt.subplots() 

# 标题
plt.title("二维空间结构示意图")

# 绘制点
x = 0
y = 0
ax.scatter([-2, -3, 3, 4, -3], [2, 4, 4, -3, -1])
ax.scatter(x, y, marker='x', color='red')

# 带箭头的文字说明
plt.annotate("参考点：原点", 
        c="r", 
        weight="bold", 
        xy=(x, y),  # 被注释图形内容的位置坐标
        xytext=(x + 0.8, y + 1.2),  # 注释文本的位置坐标
        arrowprops=dict(arrowstyle="->", connectionstyle="arc3", color="r") # 被注释内容的箭头属性字典
),

# 参考线
ax.axvline(x=x, linestyle='--')
ax.axhline(y=y, linestyle='--')

# 显示图形
plt.show()
```

## 如何表示二维向量？ 

你可以把一个向量想象成平面上的一个直线箭头，从原点指向一个具体的点。

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"] # 设置默认字体为 SimHei，以便支持中文
mpl.rcParams["axes.unicode_minus"] = False # 设置为False，以便图形中负号正常显示。
fig, ax = plt.subplots() 


# 标题
plt.title("用箭头来表示二维向量")

# 设置x轴和y轴的范围
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# 绘制点
x = 0
y = 0
n = 3
m = 2
ax.scatter(x, y, marker='x', color='red')
ax.scatter(n, m, marker='x', color='red')
ax.text(x + 0.2, y - 0.2, f"({x}, {y})")
ax.text(n + 0.2, m - 0.2, f"({n}, {m})")
ax.arrow(x, y, n - 0.2, m - 0.15, head_width=0.2, head_length=0.2, fc="black", ec="black")

# 辅助虚线
ax.axvline(x=x, linestyle='--')
ax.axvline(x=n, linestyle='--')

# 显示图形
plt.show()
```

```python
# 绘制恐龙

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

mpl.rcParams["font.sans-serif"] = ["SimHei"] # 设置默认字体为 SimHei，以便支持中文
mpl.rcParams["axes.unicode_minus"] = False # 设置为False，以便图形中负号正常显示。
fig, ax = plt.subplots() 

# 标题
plt.title("示例：绘制恐龙")

# 绘制点
dino_vectors = [(6,4), (3,1), (1,2), (-1,5), (-2,5), (-3,4), (-4,4),
    (-5,3), (-5,2), (-2,2), (-5,1), (-4,0), (-2,1), (-1,0), (0,-3),
    (-1,-4), (1,-4), (2,-3), (1,-2), (3,-1), (5,1)
]

ax.scatter(0, 0, marker='x', color='red')
for item in dino_vectors:
    ax.scatter(*item, color='black')

# 绘制多边线
polygon = Polygon(dino_vectors, fill=None)
ax.add_patch(polygon)

# 显示图形
plt.show()
```
