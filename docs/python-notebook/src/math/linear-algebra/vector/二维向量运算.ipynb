# 二维向量运算

和数一样，向量也有自己的运算方式。对向量进行运算可以生成新的向量，向量运算包括：

1. 向量加法

```python
# 向量加法

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams["font.sans-serif"] = ["SimHei"] # 设置默认字体为 SimHei，以便支持中文
mpl.rcParams["axes.unicode_minus"] = False # 设置为False，以便图形中负号正常显示。
fig, ax = plt.subplots() 

# 标题
plt.title("二维向量加法")

# 设置x轴和y轴的范围
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# 绘制点
x = 0
y = 0
n = 3
m = 2
ax.arrow(x, y, n - 0.2, m - 0.15, head_width=0.2, head_length=0.2, fc="black", ec="black")

# 辅助虚线
ax.axvline(x=0, lw=3)
ax.axhline(y=0, lw=3)

# 显示图形
plt.show()
```
